6. Perform the data clustering algorithm using any Clustering algorithm

```python
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data  # Features: sepal length, sepal width, petal length, petal width
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=100)
plt.title("K MEANS CLUSTERING - IRIS DTATASET")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show
```

```python
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')
```

```python
plt.title('K-Means Clustering (Iris Dataset)')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
```

```python
true_labels = iris.target
```

```python
ari_score = adjusted_rand_score(true_labels, labels)
print(f"Adjusted Rand Index (ARI) score: {ari_score:.4f}")
```
